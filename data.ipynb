```python
# extract total races from every season starting from 2000 to 2021
season_total_races = {}

for i in range(22):
    season = 2000 + i
    url = "http://ergast.com/api/f1/{season}.json".format(season=season)
    season_request = requests.get(url=url)

    season_total_races[season] = int(season_request.json()['MRData']['total'])

season_total_races
```

```python
#functions that formats api json output into dataframe

def format_df(year, round):
    r = requests.get(url="http://ergast.com/api/f1/{year}/{round}/results.json".format(year=year, round=round))
    response = r.json()['MRData']
    season = response['RaceTable']['Races'][0]['season']
    round = response['RaceTable']['Races'][0]['round']
    raceName = response['RaceTable']['Races'][0]['raceName']
    circuitName = response['RaceTable']['Races'][0]['Circuit']['circuitName']
    result_df = pd.DataFrame(response['RaceTable']['Races'][0]['Results'])
    

    result_df['season'] = season
    result_df['round'] = round
    result_df['raceName'] = raceName
    result_df['circuitName'] = circuitName
    
    return result_df
```

```python
#get racing data for the first race of 2021 (Bahrain GP)
format_df(2021, 1)
```

```python
races_df_lst = []
for i in range(22):
    season = 2000+i
    total_races = season_total_races[season]

    for j in range(total_races):
        round = j + 1
        races_df_lst.append(format_df(season,round))

```

```python
data_final = pd.concat(races_df_lst)
```

```python
data_final['Constructor'] = data_final['Constructor'].apply(lambda x: x['constructorId'])
data_final['Driver'] = data_final['Driver'].apply(lambda x: x['driverId'])
```

```python
data_final.to_csv('f1_race_results.csv', index=False)
```
